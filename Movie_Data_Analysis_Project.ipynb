#  Movie Data Analysis Project (1980–2020)
**For Beginners — Using Python & Pandas**

In this project, we’ll explore a dataset of movies released between 1980 and 2020. We’ll use Python to clean, analyze, and visualize this data. This will help you learn how to do data analysis step-by-step.

➡️ Dataset contains columns like: Title, Director, Genre, Release Date, Duration, Rating.

```python
# Step 1: Load Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 2: Load the Dataset
df = pd.read_csv('movies_1980_2020_30k.csv')
df.head()
```

```python
# Step 3: Check for Missing Data & Clean it
print(df.isnull().sum())
# Convert 'Release Date' to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
# Drop rows with invalid dates
df = df.dropna(subset=['Release Date'])
df.dtypes
```

```python
# Step 4: Basic Statistics
print('Average Duration:', df['Duration'].mean())
print('Rating Stats:')
print(df['Rating'].describe())
print('Top Genres:')
print(df['Genre'].value_counts().head())
```

```python
# Step 5: Rating Distribution Visualization
plt.figure(figsize=(8,4))
sns.histplot(df['Rating'], bins=20, kde=True,color='lightgreen')
plt.title('Rating Distribution',color='seagreen')
plt.xlabel('Rating', color='blue')
plt.ylabel('Number of Movies',color='blue')
plt.show()
```

```python
# Step 6: Top 10 Genres Visualization
plt.figure(figsize=(10,4))
df['Genre'].value_counts().head(10).plot(kind='bar', color='lavender')
plt.title('Top 10 Movie Genres',color='red')
plt.xlabel('Genre',color='green')
plt.ylabel('Number of Movies',color='green')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 7: Movies Released Each Year
df['Year'] = df['Release Date'].dt.year
movies_per_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12,5))
movies_per_year.plot()
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year',color='blue')
plt.ylabel('Count',color='blue')
plt.grid(True)
plt.show()
```

```python
# Step 8: Top 10 Directors by Movie Count
top_directors = df['Director'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_directors.plot(kind='barh', color='purple')
plt.title('Top 10 Directors',color='blue')
plt.xlabel('Movie Count',color='purple')
plt.ylabel('Director',color='purple')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Step 9: Does Duration Affect Ratings?
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Duration', y='Rating',color='pink')
plt.title('Relation Between Duration and Rating',color='purple')
plt.xlabel('Duration (in minutes)',color='blue')
plt.ylabel('Rating',color='blue')
plt.grid(True)
plt.show()
```

```python
# Step 10: Save Cleaned Dataset
df.to_csv('cleaned_movies.csv', index=False)
```
